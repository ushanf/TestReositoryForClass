```python
import pandas as pd
Location = "datasets/titanic3.xls"
df = pd.read_excel(Location)
df.head()
```

```python
def score_to_numeric(x):
    if x=='female':
        return 1
    if x=='male':
        return 0
```

```python
df['gender_val']=df['sex'].apply(score_to_numeric)
df.head()
```

```python
df['relatives']=df['sibsp']+df['parch']
df.head()
```
